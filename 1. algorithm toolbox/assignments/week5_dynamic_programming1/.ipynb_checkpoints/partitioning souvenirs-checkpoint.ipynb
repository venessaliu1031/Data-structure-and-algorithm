```python
# Uses python3
import sys
import itertools

def partition3(A):
    n = len(A)
    total = 0
    i, j = 0, 0
      
    # calculate sum of all elements 
    for i in range(n): 
        total += A[i] 
      
    if total % 3 != 0: 
        return 0
      
    part = [[ 0 for i in range(n + 1)]  
                   for j in range(total//3 + 1)] 
      
    # initialize top row as true 
    for i in range(0, n + 1): 
        part[0][i] = 1
          
    # initialize leftmost column,  
    # except part[0][0], as 0 
    for i in range(1, total//3 + 1): 
        part[i][0] = 0
      
    # fill the partition table in 
    # bottom up manner 
    for i in range(1, total//3 + 1): 
          
        for j in range(1, n + 1): 
            part[i][j] = part[i][j - 1] 
              
            if i >= A[j - 1]: 
                part[i][j] = part[i][j] or part[i - A[j - 1]][j - 1]
            #if i == total//3 +1 and part[i][j] == 0: return 0
    onethird = part[total//3][n]
    # find the first half and check if it == 1     
    subset = []
    curri = total//3
    currj = n
    while (curri > 0 and currj >=0):
        if part[curri][currj] == part[curri][currj-1]:
            subset.append(A[currj-1])
            currj -= 1
        else: 
            
            curri -= A[currj-1]
            currj -= 1


            
    n2 = len(subset)
    i, j = 0, 0
      
    # calculate sum of all elements
      
    part2 = [[ 0 for i in range(n2 + 1)]  
                   for j in range(total//3 + 1)] 
      
    # initialize top row as true 
    for i in range(0, n2 + 1): 
        part2[0][i] = 1
          
    # initialize leftmost column,  
    # except part[0][0], as 0 
    for i in range(1, total//3 + 1): 
        part2[i][0] = 0
      
    # fill the partition table in 
    # bottom up manner 
    for i in range(1, total//3 + 1): 
          
        for j in range(1, n2 + 1): 
            part2[i][j] = part2[i][j - 1] 
              
            if i >= subset[j - 1]: 
                part2[i][j] = part2[i][j] or part2[i - subset[j - 1]][j - 1]
            
            
            
    
    twothird = part2[total//3][n2]
    
          
    return onethird and twothird

if __name__ == '__main__':
    #input = sys.stdin.read()
    #n, *A = list(map(int, input.split()))
    A = [3, 3, 2, 1, 3, 6]
    #onethird = 36; twothird = 72
    print(partition3(A))
```
